## Chocolate consumption and Nobel laureates

El dataset utilizado en esta tarea corresponde a la data extraida de una publicación llamada "Chocolate consumption and Nobel laureates" publicada en sciencedirect en 2020 (link: https://www.sciencedirect.com/science/article/pii/S2590291120300711#bib5).

El dataset consta de 4 columnas cuyos nombres son:
- Country 
- Chocolate consumption (kg/year/capita)	
- Coffee consumption (kg/year/capita)	
- Number of Nobel laureates (up to the year 2018)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler, StandardScaler 
plt.style.use("seaborn-dark")
```

```python
df =pd.read_excel('Chocolate_Nobel_laureates.xlsx')
```

```python
df.head()
```

```python
df.describe()
```

```python
Q1 = df['Number_of_Nobel_laureates'].quantile(q = 0.25)
Q3 = df['Number_of_Nobel_laureates'].quantile(q = 0.75)

rango = df['Number_of_Nobel_laureates'].max() - df['Number_of_Nobel_laureates'].min()
iqr = Q3 - Q1
print(rango,iqr)
```

```python
# Eliminamos el outlier
df_filtrado = df[(df['Number_of_Nobel_laureates'] < 200)]
```

```python
sns.pairplot(
    df_filtrado,
    x_vars=["Chocolate_consumption_kg_year_capita"],
    y_vars=["Number_of_Nobel_laureates"],
    kind="reg",
    height=5
)
plt.show()
```

```python
sns.jointplot(data=df_filtrado, x='Chocolate_consumption_kg_year_capita', y='Number_of_Nobel_laureates')
plt.show()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(df_filtrado[['Chocolate_consumption_kg_year_capita', 'Number_of_Nobel_laureates']])
```

```python
scaled.T.shape
```

```python
covariance_matrix = np.cov(scaled.T) #Datos Traspuestos
covariance_matrix
```

```python
plt.figure(figsize =(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix, 
                cbar=True,
                annot=True,
                square=True,
                annot_kws={'size':12},
                yticklabels =['Chocolate_consumption_kg_year_capita', 'Number_of_Nobel_laureates'],
                xticklabels =['Chocolate_consumption_kg_year_capita', 'Number_of_Nobel_laureates'])
```

- Existe una correlación positivas pero es débil con un valor de 0.36 respecto a la cantidad de premios Nobel y el consumo de chocolate.
- No se puede reproducir lo expuesto por Messerili, 2012 con un 0.791, que es en lo que se basa el paper utilizado para este análisis.
