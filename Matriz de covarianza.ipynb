```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler # Z-score
plt.style.use("seaborn-dark")

iris = sns.load_dataset('iris')
```

```python
#sns.pairplot(iris, hue ='species')
sns.pairplot(iris, height=5, aspect=0.8, kind='reg')
plt.show()
```

```python
scaler = StandardScaler()
scaled = scaler.fit_transform(iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
```

```python
# Standard Scaler es la normalizacin que tiene en cuenta restar por promedio
# y dividir por la desviacin estandar
```

```python
scaled.T.shape #transponer el scaled
```

```python
covariance_matrix = np.cov(scaled.T) #Datos Traspuestos
covariance_matrix
```

```python
# las variables con mas valor son aquellas que estan relacionadas con ellas mismas
```

```python
plt.figure(figsize =(10,10))
sns.set(font_scale=1.5)
hm = sns.heatmap(covariance_matrix, 
                cbar=True,
                annot=True,
                square=True,
                annot_kws={'size':12},
                yticklabels =['sepal_length','sepal_width', 'petal_length', 'petal_width'],
                xticklabels =['sepal_length','sepal_width', 'petal_length', 'petal_width'])
```

<h5 style ='text-align:center'>Covarianza </h5>
-Cuando entre las dos variables hay una <span style='font-weight:bold; color:#AED581'>relacin directa</span>, la covarianza da un valor positivo.

-Cuando entre las dos variables hay una <span style='font-weight:bold; color:#AED581'>relacin inversa</span>, la covarianza da un valor negativo.

-Cuando entre las dos variables no hay relacin, la covarianza da un valor enotorno a cero

```python
sns.heatmap(iris.corr(), annot=True)
plt.show()
```
