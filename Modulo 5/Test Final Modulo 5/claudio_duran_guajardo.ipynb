```python
from sklearn.datasets import load_wine
import pandas as pd
```

```python
vinos = load_wine()
vinos
print(vinos['DESCR'])
```

```python
# Nombres de las variables del dataset
vinos.feature_names
```

```python
# Creaci√≥n del DataFrame utilizando los nombres 
vinos_DF = pd.DataFrame(vinos.data, columns = vinos.feature_names)
vinos_DF['objetivo'] = vinos.target
vinos_DF
```

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
# Se crea el √°rbol de Decisi√≥n
arbol = tree.DecisionTreeClassifier()
```

```python
X_train,X_test,y_train,y_test = train_test_split(vinos_DF.drop("objetivo", axis =1),
                                                vinos_DF.objetivo,test_size = 0.3, random_state=42)
```

```python
# Entrenamos al modelo
arbol.fit(X_train,y_train)
```

```python
predicciones = arbol.predict(X_test)
predicciones
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
def accuracy_cdg(modelo,X,y):
    preds = modelo.predict(X)
    return accuracy_score(y, preds)

def precision_cdg(modelo,X,y):
    preds = modelo.predict(X)
    return precision_score(y, preds, average="weighted")

def sensibilidad_cdg(modelo,X,y):
    preds = modelo.predict(X)
    return recall_score(y, preds, average="weighted")

def f1_score_cdg(modelo,X,y):
    preds = modelo.predict(X)
    return f1_score(y, preds, average="weighted")

```

```python
# Evaluamos el modelo con la exactitud
cvs_accuracy = cross_val_score(arbol, X_test,y_test,scoring=accuracy_cdg,cv=10).mean()
cvs_precision = cross_val_score(arbol, X_test,y_test,scoring=precision_cdg,cv=10).mean()
cvs_sensibilidad = cross_val_score(arbol, X_test,y_test,scoring=sensibilidad_cdg,cv=10).mean()
cvs_f1_score = cross_val_score(arbol, X_test,y_test,scoring=f1_score_cdg,cv=10).mean()
cvs_evaluacion_modelo  = {"accuracy":cvs_accuracy,
                          "precision":cvs_precision,
                          "sensibilidad":cvs_sensibilidad,
                          "f1 score":cvs_f1_score}
cvs_evaluacion_modelo
```

- El promedio de exactitud (Accuracy) en evaluaci√≥n Cruzada es superior al 90%
- El promedio de Precision en evaluaci√≥n Cruzada es superior al 90%
- El promedio de Sensibilidad en evaluaci√≥n Cruzada es superior al 90%
- El promedio de F1 Scores en evaluaci√≥n Cruzada es un poco inferior al 90%
