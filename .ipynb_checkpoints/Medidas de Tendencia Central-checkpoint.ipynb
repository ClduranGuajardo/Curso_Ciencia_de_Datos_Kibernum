## Media (Promedio), Mediana y la Moda

```python
import pandas as pd
pd.options.display.max_columns = None
import matplotlib.pyplot as plt
plt.style.use("seaborn-dark")
```

```python
df =pd.read_csv('cars.csv') # Data de automoviles usados
```

```python
df.head()
```

```python
df['price_usd'].mean()
```

```python
# Nos interesa analizar los atipicos también es decir cual es el auto con mayor precio
# cual de todos es el barato
```

```python
df['price_usd'].median()
# con la mediana puedo identificar que existen autos costosos a pesar de ser usados
# el promedio esta lejado a la mediana
```

```python
# Vamos a graficar la columna price_usd
df['price_usd'].plot.hist(bins = 20, edgecolor ='black') # 20 segmentaciones o intervalos de valores
plt.show()
```

```python
#la mayoria de ls autos estan entre 0 a 10000, pero tenemos varios casos atipicos
```

```python
import seaborn as sns
```

```python
sns.displot(df, x='price_usd', hue ='manufacturer_name')
plt.show()
```

```python
#pasamos a una visualización mas sencilla
sns.displot(df, x ='price_usd', hue = 'engine_type', multiple ='stack')
plt.show()
```

```python
# Aun sigo sin poder ver los autos electricos, puedo intuir que hay pocos autos electricos
# Conteo tambien es un estadistico descriptivo
# Calculamos cuantos autos hay por cada tipo de maquina de acuerdo a su motor(engine_type)?
```

```python
df.groupby('engine_type').count()
```

```python
# Solo hay 10 autos electricos
# La estadistica descriptiva no solo son numeros sino que también visualizaciones
```

```python
# busquemos un auto que sea de marca Audiy de la clase Q7
Q7_df = df[(df['manufacturer_name'] == 'Audi')&(df['model_name'] == 'Q7')]
```

```python
#Q7_df.iloc[:]
Q7_df
```

```python
# Modelo por año
sns.histplot(Q7_df, x='price_usd', hue = 'year_produced')
plt.show()
```

Observaciones:
- La mayoria de los autos Q7 que estan en este catalogo tienden a tener precios bajos entre 10000 a 20000
- Los colores claros hacen referencia a los modelos más antiguos del 2006, 2007, 2008
- El auto más antiguo va depreciando su valor

```python
#Busquemos los autos Marca KIA y Modelo Carnival
Carnival_df = df[(df['manufacturer_name'] == 'Kia')&(df['model_name'] == 'Carnival')]
Carnival_df.shape
```

```python
Carnival_df
```

```python
# Modelo por año
sns.histplot(Carnival_df, x='price_usd', hue = 'year_produced')
plt.show()
```

Observaciones:
- La mayoria de los autos Carnival que estan en este catalogo tienden a tener precios bajos entre 1000 a 7000
- Los colores claros hacen referencia a los modelos más antiguos del 1999, 2000, 2001
- El auto más antiguo va depreciando su valor
- Existe una anomalia entorno a los valores 4000 y 5000 donde modelos más antiguos tienen un valor mayor que autos más nuevos, en cuanto a años de producción
